Un  intercambiador de calor en paralelo tiene una conductancia total de $UA=10 $ W/K. El fluido caliente ingresa a $T_{c,i} = 400$ K y tiene una capacidad térmica $C_c= 10$ W/K. El fluido frío ingresa a $T_{f,i} = 300$ K and tiene una capacidad térmica de $C_f = 5 $W/K.
1. Determinar el número de unidades de transferencia (*NUT*), la eficiencia $\varepsilon$, el flujo de calor $\dot q$ y las temperaturas de salida $T_{c,s}$ y $T_{f,s}$ para el intercambiador.
2. Realizar un esquema de la distribución de temperaturas en el intercambiador.
3. Realizar un esquema de la distribución de temperaturas en el intercambiador si la conductancia es muy grande.

4. Realizar el esquema que corresponde a los cambios de temperaturas a medida de que la conductancia $UA$ varía, manteniendo las demás cantidades constantes. $0<UA<\infty$


```python
Tf1,Tc1 = [300.,400.]
delta_Tmax = Tc1-Tf1
UA = 10.
Cf,Cc = [5.,10.]
Cmin = np.min([Cf,Cc])
Cmax = np.max([Cf,Cc])
NUT1 = ht.NTU_from_UA(UA=UA, Cmin=Cmin)
Cr = Cmin/Cmax
eff1 = ht.effectiveness_from_NTU(NTU=NUT1, Cr=Cr, subtype='parallel')
q1 = Cmin*delta_Tmax*eff1
Tf2 = q1/Cf+Tf1
Tc2 = Tc1-q1/Cc
```

```python
L = 1
x = np.linspace(0,L,100)
NUTx = UA*x/L
effx = []
for i,NUTxi in enumerate(NUTx):
    effx.append(ht.effectiveness_from_NTU(NTU=NUTxi, Cr=Cr, subtype='parallel'))
effx = np.asarray(effx)
qx = Cmin*delta_Tmax*effx

Tfx = qx/Cf+Tf1
Tcx = -qx/Cc+Tc1
```

```python
fig0,ax0 = plt.subplots(figsize=(4,4))
ax0.plot(x,Tfx)
ax0.plot(x,Tcx)
```
