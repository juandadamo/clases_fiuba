```python
%matplotlib widget 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tikzplotlib import save as savetikz
from IPython.display import Latex
```

```python
t = pd.date_range(
    start='2023-11-01',
    end='2024-09-01',
    periods=11
)
inflacion = np.array([0,25.5,51.35,71.33,90.18,106.91,115.6,125.52,134.61,143.53,152.54])
salario   = np.array([0,5.5,5.5,22.39,37.07,48.04,61.36,67.82,80.40,85.81,89.42])

inflacion = inflacion+100
salario = salario+100

dict_salario = {'tiempo':t,'inflacion':inflacion,'salario':salario}
data_out = pd.DataFrame(dict_salario)
data_out.to_csv('salarios_sep2024.csv')
```

```python
fig,ax = plt.subplots(figsize=(7,3))
ax.plot(t,inflacion,'o',fillstyle='none',markeredgewidth=2)
ax.plot(t,salario,'s',fillstyle='none',markeredgewidth=2)
ticks_to_use = t
# Set format of labels (note year not excluded as requested)
labels = [ i.strftime('%b %Y').replace('202','2') for i in ticks_to_use ]

# Now set the ticks and labels
ax.set_xticks(ticks_to_use)
ax.set_xticklabels(labels,fontsize=10)
ax.grid()
savetikz('salarios_sep2024.tikz')
```

```python
dt=1
Acum_inflacion = np.trapz(inflacion)*dt
Acum_salario   = np.trapz(salario)*dt
Dif_acum = (Acum_inflacion-Acum_salario)/salario[-1]
```

```python
display(Latex(f'Lo acumulado por inflacin desde Noviembre de 2023 es: {Acum_inflacion:.0f}'))
display(Latex(f'Lo acumulado por salario desde Noviembre de 2023 es: {Acum_salario:.0f}'))
display(Latex(f'La diferencia acumulada es de  {Dif_acum:.1f} salarios de septiembre de 2024.'))
```
