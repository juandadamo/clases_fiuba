```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
mx,my = np.meshgrid(x,y)
mr = (mx**2+my**2)**.5

e=0.6

TF = 10
TC = 30
R1 = 1
R2 = 4
R3 = 4+e
Tr = TC+(TC-TF)*np.log(mr/R1)  /(np.log(R1/R2))
Tr2 = TC+(TC-TF)*np.log(mr/R2)  /(np.log(R2/R3))

Tr[mr<R1] = TC
Tr[mr>R2] = Tr2[mr>R2]
Tr[mr>R3] = 0
lam1 = 50
lam2 = 1
Q1 = lam1*2*np.pi/ (np.log(R1/R2)) * (TC-TF)
Q2 = lam2*2*np.pi/ (np.log(R2/R3)) * (TC-TF)
```

```python
fig0,ax0 = plt.subplots(figsize=(3,3))
ax0.contourf(x,y,Tr)
```
