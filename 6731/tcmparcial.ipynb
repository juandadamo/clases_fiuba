# Problema Radiacion

Problema 2
Se requiere colocar un panel circular de calentamiento A1 en el techo de una habitación para
calefaccionar una pequeña superficie A2 ubicada en forma perpendicular y centrada con el eje del
disco, a 2m de distancia. La misma tiene características de cuerpo negro. El panel está recubierto
con una pintura blanca mate que le proporciona una emitancia de 0.88. Como criterio de diseño
se requiere que la superficie A2 alcance una temperatura de 305K, con A1 a 355K; y tanto el aire
del ambiente como las paredes que lo determinan A3 se encuentran a 290K. Se puede suponer
que el coeficiente de transferencia de calor convectiva entre A2 y el aire es de hc =4W/m2 K. Se
pide:
1. Plantear el balance de energía sobre la supuerficie A2.
2. Hallar el factor de forma F21 (Ayuda: No sale de tabla ni hace falta conocer la superficie
A2 )
3. Determinar el Diámetro de la superficie A1.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
```

```python
h = 4
L = 2
T3= 290
T2 = 305
eps1 = 0.8
sigma = constants.Stefan_Boltzmann


qconv = h * (T2-T3)
qrad = qconv
Ebs = (T1**4-T2**4)*sigma
aux = (qrad / Ebs)**-1-(1-eps1)/eps1
F21 = 1/aux
```

# Problema aleta

Se quiere usar un termómetro de mercurio para medir la temperatura de un gas caliente que
circula por el interior de un conducto de gran diámetro. Para una correcta medición, se perfora
la pared del conducto y se inserta en dirección radial un tubo de acero inoxidable de diámetro
D =7 mm, espesor t =0.7mm y largo L =82mm. Uno de sus extremos se suelda a la pared del
tubo (lo que garantiza un perfecto contacto térmico) y el otro se cierra con el fin de evitar fugas
del gas a través del tubo. Por el interior de este se inserta el termómetro hasta obtener un contacto
íntimo con el tubo en el extremo cerrado (x = L). Si se sabe que la temperatura de la pared
del conducto es T0 = 240◦ C y el termómetro indica TL = 318◦ C, ¿Cuál es la temperatura del
gas? Considerar conductividad térmica del tubo k = 15W/m K, un coeficiente de transferencia
de calor por convección entre el gas y el tubo de 30W/m2 K. Utilizar modelo de aleta aguja de
extremo aislado.

```python
D=7e-3
L = 82e-3
t = 0.7e-3
Tb = 240
TL = 318
k = 15
h = 30
```

```python
P = D*np.pi
St = D**2*np.pi/4 - (D-t)**2*np.pi/4
beta  = np.sqrt(h*P/(k*St))
betaL = beta*L
```

```python
Tinf =(TL - Tb/np.cosh(betaL))*(1-1/np.cosh(betaL))**-1
print(Tinf)
```
