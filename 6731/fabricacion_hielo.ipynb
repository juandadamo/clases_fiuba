En cierto proceso de fabricación  de hielo, el agua a 5$^\circ$C fluye sobre una placa enfriada por un refrigerante a -10$^\circ$C. Cuando la capa de hielo alcanza los 2mm de espesor, se raspa para quitarla de la placa y el proceso se repite. Calcule el intervalo de tiempo que transcurre entre dos raspaduras. Tome un valor de 900 W/m$^2$ K para el coeficiente global de transferencia de calor del refrigerante a la superficie de contacto entre la placa y el hielo. La entalpia de fusión del hielo es de 335 kJ/kg.

Considerando que el número de Jakob para la formación de hielo es pequeño, se puede suponer que la conducción es cuasiestacionaria. El balance en la interfaz hielo-agua establece que:
\begin{equation}
\left.\lambda _1\nabla T_1\right|_s=\left.\lambda _2\nabla T_2\right|_u+\dot{m}h_{fs} 
\end{equation}

El miembro de la izquierda puede despreciarse, si $Ja\ll 1$. El término de la caída de temperatura por conducción resulta de la suma de la caída desde la placa hasta el hielo más la caída en el hielo.:$\lambda_2\left.\partial T / \partial z\right|= (T_s - T_0) / \lbrace s/k + 1/U\rbrace$. La masa que solidifica se estima según $\dot m = -\rho ds/dt$. Luego,

$$\rho h_{fs} ds/dt =\frac{T_s - T_0}{s/k + 1/U}$$

Si para $t=0$, $s=0$:

$$\frac{T_s - T_0}{\rho h_{fs}} dt = {s/k + 1/U} ds$$

Reconozcamos el número de Jakob en el miembro izquierdo. Integrando:

$$\frac{T_s - T_0}{\rho h_{fs}} t = s^2/2k + s/U $$

con los datos del problema,


```python
Ts,T0,rho,k,hfs,U,s=[0,-10,910,2.22,335e3,900,2e-3]

print ('t=%.4f seg.'%(rho*hfs*(s**2/(2*k)+s/U)/(Ts-T0)))

```
