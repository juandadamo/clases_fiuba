# **Apunte: Funciones de Bessel y su Aplicación**  

Las funciones de Bessel surgen como soluciones de ecuaciones diferenciales en problemas con simetría cilíndrica y esférica, desempeñando un papel clave en la física matemática y la ingeniería.  

## **Funciones de Bessel estándar**  

La ecuación diferencial de Bessel:  

$$
\frac{d^2y}{dx^2} + \frac{1}{x} \frac{dy}{dx} + \left(a^2 - \frac{n^2}{x^2} \right)y = 0  
$$

tiene como solución general:  

$$
y = A J_n(ax) + B Y_n(ax),
$$

donde $ J_n(x) $ y $ Y_n(x) $ son las funciones de Bessel de primera y segunda especie, respectivamente.  

Las funciones de Bessel de primera especie $ J_n(x) $ están definidas por la serie de potencias:  

$$
J_n(x) = \sum_{k=0}^{\infty} \frac{(-1)^k}{k! \, \Gamma(k+n+1)} \left(\frac{x}{2}\right)^{2k+n},
$$

mientras que las funciones de Bessel de segunda especie $ Y_n(x) $ se expresan en términos de $J_n(x) $ y la función gamma $ \Gamma(x) $:  

$$
Y_n(x) = \frac{J_n(x) \cos(n\pi) - J_{-n}(x)}{\sin(n\pi)}.
$$

## **Funciones de Bessel modificadas**  

La ecuación diferencial modificada de Bessel:  

$$
\frac{d^2y}{dx^2} + \frac{1}{x} \frac{dy}{dx} + \left(-a^2 - \frac{n^2}{x^2} \right)y = 0  
$$

admite la solución:  

$$
y = C I_n(ax) + D K_n(ax),
$$

donde $I_n(x)$ y $K_n(x) $ son las funciones de Bessel modificadas de primera y segunda especie, respectivamente.  

Las funciones de Bessel modificadas de primera especie $I_n(x) $ están definidas por:  

$$
I_n(x) = \sum_{k=0}^{\infty} \frac{1}{k! \, \Gamma(k+n+1)} \left(\frac{x}{2}\right)^{2k+n},
$$

mientras que las funciones de Bessel modificadas de segunda especie $ K_n(x) $ se expresan en términos de $ I_n(x) $:  

$$
K_n(x) = \frac{\pi}{2} \frac{I_{-n}(x) - I_n(x)}{\sin(n\pi)}.
$$

Las funciones $ I_n(x) $ y $ K_n(x) $ son especialmente útiles en problemas de difusión y propagación en medios sin oscilaciones, como el flujo de calor en cilindros y la propagación de ondas evanescentes.

## **Funciones de Bessel Escaladas** 

Una aclaración importante es que en la bibliografía es posible encontrar tablas donde las funciones se encuentran escaladas exponencialmente, entonces se definen las funciones de bessel escaladas exponencialmente de la manera:
$$
B_{ne}(n,z)=B_{n}(n,z)*e^{z}
$$

## **Uso de Librería Scipy**

La librería Scipy de python tiene un módulo [scipy.special](https://docs.scipy.org/doc/scipy/reference/special.html) en el cual se presentan las funciones de bessel para obtener sus resultados para determinados valores de x. También existen las versiones rápidas de dichas funciones para aquellas cuyo orden $n$ es cero o uno. Para mayor profundidad visitar la página de scipy.special y el Apéndice B.3 del Mills


```python
import scipy.special as sc
import numpy as np

#Función de Bessel de primera especie
def Jn(n,r):
    #n = el orden de la funcion
    #r=variable
    j=sc.jn(n,r)
    return j

#Función de Bessel de segunda especie
def Yn(n,r):
    #n = el orden de la funcion
    #r=variable
    y=sc.yn(n,r)
    return y

#Funcion de bessel Modificada de primera especie Jn   
def Kn(n,r):
    #n = el orden de la funcion
    #r=variable
    k=sc.kv(n,r)
    return k

#Funcion de bessel Modificada de segunda especie Yn
def In(n,r,a):
    #n = el orden de la funcion
    #r=variable
    i=sc.iv(n,a*r)
    return i


#Grafico funciones de Bessel de primera especie para n=0,1
import matplotlib.pyplot as plt
r=np.linspace(0,20,100)
plt.plot(r,Jn(0,r),label='J0(r)')
plt.plot(r,Jn(1,r),label='J1(r)')
plt.legend()
plt.title('Funciones de Bessel de primera especie')
plt.xlabel('r')
plt.ylabel('Jn(r)')
plt.grid()
plt.show()


#Grafico funciones de Bessel de segunda especie para n=0,1
import matplotlib.pyplot as plt
r=np.linspace(0,20,100)
plt.plot(r,Yn(0,r),label='Y0(r)')
plt.plot(r,Yn(1,r),label='Y1(r)')
plt.legend()
plt.title('Funciones de Bessel de segunda especie')
plt.xlabel('r')
plt.ylabel('Yn(r)')
plt.grid()
plt.show()
```
