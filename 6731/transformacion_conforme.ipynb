```python
%matplotlib widget
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = 1
a2 = 0.5
d1 = 2
d2 = 1.5

d = d1-d2
a = 3
c1 = np.sqrt(d1**2-a1**2)



x = np.linspace(-100,100,200)
y = np.linspace(-100,100,200)
X,Y = np.meshgrid(x,y)
U = X**2+(Y-d)**2-(d**2+c**2) 
V = (X-d)**2+Y**2 - (d**2-c**2)

fig,ax = plt.subplots()
ax.contour(X,Y,U,colors='k')
ax.contour(X,Y,V,colors='tab:blue')
```
