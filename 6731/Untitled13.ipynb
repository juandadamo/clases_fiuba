```python
%matplotlib widget
import numpy as np
from scipy.optimize import bisect,brentq,fsolve
import matplotlib.pyplot as plt
```

```python


def cot_x(x, A):
    return 1/np.tan(x) - x/A

def find_roots(n, A):
    roots = []
    x0 = 0.001  # Valor inicial para comenzar la búsqueda de la raíz
    while len(roots) < n:
        root = bisect(cot_x, x0, x0+np.pi, args=(A,))
        roots.append(root)
        x0 = root + np.pi/2  # Avanzar al siguiente intervalo
    return roots


fcot = lambda x : x * np.cos(x-4)
```

```python
A = 2
raices = brentq(cot_x, 0, 15, args=(A), full_output=True, disp=False)
```

```python
x = np.linspace(1,10,300)
fig,ax = plt.subplots()
ax.plot(x,cot_x(x,A))
ax.plot(raices,0,'ro')
```
