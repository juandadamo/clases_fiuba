```python
U1 = np.linspace(0,80,100)
U = U1/3.6
D = 0.8
L = 1
A = L*D
nu = 1.5e-5
a = 2.2e-5
k = 0.02625
Re = U*D/nu
Pr= nu/a
Th = 37
Tinf = 10


Potencia = 50*U+100

Nusselt = 0.19*Re**0.63*Pr**0.36
h = Nusselt*k/D
Q = h*A*(Th-Tinf)
```
