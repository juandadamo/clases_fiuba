uniforme de $T_W =100^\circ$ C y separa 2 fluidos. A la izquierda, vapor de agua saturado que se condensa en forma de

intercambio. Una parte de la curva de ebullición para el líquido-X se muestra a continuación. El vapor saturado y el
líquido-X saturado se suministran al sistema, mientras que el condensado de agua y vapor-X se eliminan por
medios que no se muestran en el dibujo (Presión constante a ambos fluidos). A presión de 1bar, el fluido-X tiene
una temperatura de saturación y calor latente de T_{sat} =80$^\circ$C y $h_{fg} =$700kJ/kg, respectivamente. Asumiendo estado
estacionario, y que las resistencia a la conducción en la placa es despreciable, se pide:

    (a) Determinar el flujo de calor que se establece.
    (b) Determine la temperatura y presión de saturación para el vapor, suponiendo que ocurre condensación en
    película. LEER LAS PROPIEDADES DEL VAPOR A $T_W$ Y NO HACER NINGUNA CORRECION SOBRE ESTAS.
    (c) Estime los flujos másicos evaporados y condensados

    (e) Realice el circuito eléctrico equivalente general, indicando:

    - SOLO PARA ESTE PUNTO, ASUMA QUE NO CONOCE LOS VALORES RELATIVOS ENTRE LAS RESISTENCIAS.
    (f) ¿Es correcto el régimen supuesto en (b) para la película de condensado? Justifique o indique la estrategia para la corrección.
    (g) ¿Cómo corrigiría las propiedades térmicas del agua debidio a la variación de la temperatura en la película? (NO HACER LAS CUENTAS, SOLO INDICAR LA ESTRATEGIA)

```python
data_curva = pd.read_csv('data_ebullcion_problema.csv', delimiter=';',decimal=",",skiprows=0)
deltaT,q = data_curva.to_numpy().T

fig0,ax0 = plt.subplots(figsize=(4,4))
f_ebu = interpolate.interp1d(deltaT,q, kind='cubic')
ax0.loglog(deltaT,q,'s')
ax0.loglog(deltaT,f_ebu(deltaT))
ax0.grid()
```

```python
#Dimensiones de la placa
L = 2
A = L**2
T_w = 100 
T_satX = 80
h_fgX = 700e3
g =9.8
sat_liquid=IAPWS97(T=T_w+273, x=0)  
sat_vapor=IAPWS97(T=T_w+273, x=1)  

rho_l = cp.CoolProp.PropsSI('D','T',273+T_w,'Q',0,'water')
rho_v = cp.CoolProp.PropsSI('D','T',273+T_w,'Q',1,'water')
k_l = cp.CoolProp.PropsSI('L','T',273+T_w,'Q',0,'water')
mu_l = cp.CoolProp.PropsSI('V','T',273+T_w,'Q',0,'water')
Cp_l = cp.CoolProp.PropsSI('C','T',273+T_w,'Q',0,'water')
Pr_l = cp.CoolProp.PropsSI('Prandtl','T',273+T_w,'Q',0,'water')
h_l = cp.CoolProp.PropsSI('H','T',273+T_w,'Q',0,'water')
h_v = cp.CoolProp.PropsSI('H','T',273+T_w,'Q',1,'water')
```

```python
h_fgl = (h_v-h_l)
h_fgl
```

```python
Delta_T = T_w-T_satX
q1 = 50e3
q = f_ebu(Delta_T)
Q = q*A
#print(q1,q)
```

```python
Q
```

```python
#h_e = ht.condensation.Nusselt_laminar(T_vapor_sat, T_we, rho_vap, rho_liq, k_agua, 
                                #nu_liq*rho_liq, h_fg2, De*5, angle=90.0)
h1a = ht.condensation.Nusselt_laminar(1, 0, rho_v, rho_l, k_l, 
                                mu_l, h_fgl, L, angle=90.0)
h1 = 0.943*(rho_l*(rho_l-rho_v)*g*h_fgl*k_l**3/(mu_l*L))**(1/4)

T_sat  = T_w+(q/h1)**(4/3)  
display(Latex(f'$h_{{cond}}={h1:.1f}$watt'))#(h1,h1a)
display(Latex(f'$T_{{sat}}={T_sat:.1f}^\circ$C'))#(T_sat)
```

```python

pres_sat = cp.CoolProp.PropsSI('P','T',273+T_sat,'Q',1,'water')
print('presion saturacion %3.1f kPa'%pres_sat)

m_cond = Q/h_fgl
m_ebu = Q/h_fgX

print("caudal másico condensado: m_cond=%3.2g"%m_cond)
print('caudal másico evaporado: m_ebu=%3.2g'%m_ebu)

h_ebu = q/Delta_T
h_cond = h1 / (T_sat-T_w)**(1/4)
print("coeff cond h_cond = %3.1f"%h_cond)
print('coeff ebullición: h_ebu=%3.2f'%h_ebu)
```
