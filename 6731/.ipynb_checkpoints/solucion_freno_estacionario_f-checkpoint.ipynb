```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
r, beta, alpha = sp.symbols('r beta alpha')
T1 = sp.Function('T1')
T2 = sp.Function('T2')
phi = sp.Function('phi')
varphi = sp.Function('varphi')
from IPython.display import display, Math,Latex
def round_expr(expr, num_digits):
    return expr.xreplace({n : round(n, num_digits) for n in expr.atoms(sp.Number)})
```

```python
b = 5 *1e-3
N = 3600 
omega = N*1/60
mu = 0.15 
P = 1 *1e6
k = 75 
Rd = 4.0 *1e-2
Rp = 3.0 *1e-2
h = 20+1500 *(omega/100)**1.25

alpha_i = P*mu*omega/b/k
beta_i = (2*h/k)**0.5
Tinf = 20
```

```python
equation2 = 2*sp.pi*sp.Derivative((r*sp.Derivative(T2(r),r,evaluate=False)),r)+4*sp.pi*r**2*alpha 
display(Latex(r'Para $r>R_p \\ \displaystyle'+ sp.latex(equation2)+'=0$'))
display(Latex(r'$\displaystyle\alpha=\frac{p\mu\omega}{k b}$'))
```

```python
T2f = sp.dsolve(equation2).args[1]
display(T2f)


#condicion de borde aislado
display(Latex(r'Condicion de borde aislado $\displaystyle\left.\frac{dT_2}{dr}\right|_{r=R_d}=0$'))
C2i = sp.solve(T2f.diff('r').subs('r',Rd).subs('alpha',alpha_i))[0]
T2f = T2f.subs('C2',C2i)
display (T2f)
```

Balance térmico en disco en contacto con pastilla de freno
$$\frac{d}{dr}(-k\frac{dT}{dr}2\pi r b)dr = 4\pi r^2p\mu\omega dr$$

Para la región $r<R_p$:

\begin{eqnarray}
\dot q|_{r+dr}d\theta(r+dr) - \dot q|_r d\theta r& =& 2h dr (rd\theta) \Delta T\\
\end{eqnarray}
$\dot q|_{r+dr} =  \dot q|_r + \partial \dot q/\partial r|_r dr $

\begin{eqnarray}
(\dot q|_r dr+ \partial \dot q/\partial r|_r r )   &=& 2h dr \Delta T \\
\partial/\partial r(r   \dot q_r)  dr &=&  2h r dr \Delta T
\end{eqnarray}
Sea $\varphi = T-T_\infty$ y $\beta^2 = 2h/k$
\begin{equation}
\frac{\partial}{\partial r}\left(r \frac{\partial \varphi}{\partial r}\right) = \beta^2 r \varphi
\end{equation}

Para $\varphi = T - T_\infty$.

```python
equation1 = sp.Derivative((r*sp.Derivative(varphi(r),r,evaluate=False)),r)+beta**2*r*varphi(r) 
display(Latex(r'La ecuación $\displaystyle'+ sp.latex(equation1)+'=0$'))


phi1 = sp.dsolve(equation1).args[1]
phi1 = phi1.subs('C1','C3').subs('C2','C4')
display(Latex(r'Tiene solución general, $\varphi (r) = '+sp.latex(phi1)+'$'))
display(Latex(r'Solución finita en $r=0, \Longrightarrow \varphi(r=0)='+sp.latex(phi1.subs('r',0))+'$'))
display(Latex(r'$\Longrightarrow \varphi(r)='+sp.latex(phi1.subs('C4',0))+'$'))
phi1 = phi1.subs('C4',0)
display(Latex(r'Calor en $\displaystyle r=R_p,\\ \displaystyle-k\frac{d\varphi}{dr}=-k\frac{dT_2}{dr}$'))

C3i = sp.solve(T2f.diff(r).subs('r',Rp).subs('alpha',alpha_i)-phi1.diff(r).subs('r',Rp).subs('beta',beta_i**2),'C3')[0]
display(Latex('$C3='+sp.latex(round_expr(C3i,2))+'$'))
phi1 = phi1.subs('C3',C3i)
T1 = phi1 + Tinf
display(Latex(r'$T_1 (r)=\varphi+T_\infty='+sp.latex(round_expr(T1,2))+'$'))

display(Latex(r'Temperaturas en $\displaystyle r=R_p,\\T_1=T_2$'))

C1i = sp.solve(T2f.subs('r',Rp).subs('alpha',alpha_i)-T1.subs('r',Rp).subs('beta',beta_i**2),'C1')[0]
T2f = T2f.subs('C1',C1i).subs('alpha',alpha_i)
display(Latex('$C1='+sp.latex(round_expr(C1i,2))+'$'))
T1 = T1.subs('beta',beta_i**2)
```
