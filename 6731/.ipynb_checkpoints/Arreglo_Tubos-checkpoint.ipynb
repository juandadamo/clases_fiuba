```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#datos geometr√≠a

Dext = 60.3e-3
esp  = 2e-3
Dint = Dext-2*esp
Tubo_area_ext = Dext**2/4*np.pi
Ntubos = 33
Ltubos = 2.68
Ntubos = np.int(Ntubos/4)*4
paso_tubos = Dext*4/3
carcasa_huelgo = 20e-3  #Cao pag 142 10mm
nfilas = 4  #considero este dato fijo para este intercambiador

ncolumnas = Ntubos / nfilas
```

```python
carcasa_xmin,carcasa_ymin,carcasa_xmax,carcasa_ymax = [0,0,paso_tubos*(ncolumnas-1)+carcasa_huelgo*2+Dext,
                                                       paso_tubos*(nfilas-1)+carcasa_huelgo*2+Dext]

coord_tubo = []
for filai in np.arange(nfilas):
    yi = filai*paso_tubos+carcasa_huelgo+Dext/2
    for columnai in np.arange(ncolumnas):
        xi = columnai*paso_tubos+carcasa_huelgo+Dext/2
        coord_tubo.append([xi,yi])

carcasa_area = carcasa_xmax*carcasa_ymax
Area_pasaje_aire = carcasa_area-Ntubos*Tubo_area_ext        

fig0,ax0  = plt.subplots()

for i,coord_tuboi in enumerate(coord_tubo):
    xi,yi = coord_tuboi
    ax0.add_patch(plt.Circle((xi, yi), Dext/2, color='k',fill=False))
ax0.set_aspect('equal')  
lin, = ax0.plot([carcasa_xmin,carcasa_xmax],[carcasa_ymin,carcasa_ymin],linewidth=3)
color_c = lin.get_color()
ax0.plot([carcasa_xmin,carcasa_xmax],[carcasa_ymax,carcasa_ymax],linewidth=3,color=color_c)
ax0.plot([carcasa_xmin,carcasa_xmin],[carcasa_ymin,carcasa_ymax],linewidth=3,color=color_c)
ax0.plot([carcasa_xmax,carcasa_xmax],[carcasa_ymin,carcasa_ymax],linewidth=3,color=color_c)
ax0.set_axis_off()
```

```python
Area_pasaje = (carcasa_ymax-Dext*nfilas)*Ltubos
print(Area_pasaje)
```
