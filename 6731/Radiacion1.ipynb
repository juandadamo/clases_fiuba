```python
%matplotlib notebook
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
import ht

from IPython.display import display, Math,Latex
def round_expr(expr, num_digits):
    return expr.xreplace({n : round(n, num_digits) for n in expr.atoms(sp.Number)})
```

```python
long_onda_sp, E_sp, uncertainties = ht.radiation.solar_spectrum()
T_sol = 5778
Temps =np.array([T_sol,T_sol*.7,T_sol/10,25])
E_cn = np.zeros((len( Temps),len(long_onda_sp)))
for j, Temp_i in enumerate(Temps):
    for i,long in enumerate(long_onda_sp):
        E_cn[j,i] = ht.radiation.blackbody_spectral_radiance(Temp_i,long)
```

```python
angulo_solido = 6.8e-5 # en estereo radianes
```

```python
fig,ax = plt.subplots()
ax.plot(long_onda_sp,E_sp)
ax.plot(long_onda_sp,E_cn[0]*6.8E-5)
ax.plot(long_onda_sp,E_cn[1]*6.8E-5)
```
